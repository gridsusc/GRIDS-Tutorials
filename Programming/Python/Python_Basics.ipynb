## 5.3 Dictionaries

### 5.3.1 Creating Dictionaries and accessing elements

Dictionaries are unordered collections of items(values) mapped with a key. The key should be unique and should be immutable(string, tuple, constants)

```python
d = {'cat': 'cute', 'dog': 'furry'}  # Create a new dictionary with some data {key:value}
print(d['cat'])       # Get an entry from a dictionary; prints "cute"
print('cat' in d)     # Check if a dictionary has a given key; prints "True"
print(type(d))
```

```python
d['fish'] = 'wet'    # Set an entry in a dictionary
print(d['fish'])      # Prints "wet"
```

```python
try:
    print(d['monkey'])  # KeyError: 'monkey' not a key of d
except KeyError as e:
    print(f"Unable to access key={e} in dictionary")
```

```python
d = dict([(1, 'one'), (2, 'two')]) #create a dictionary from tuples
print(d)
print(type(d))
```

# 6 Looping Structure & Comprehensions

## 6.1 Loops

You can loop over the elements of a list like this:

```python
animals_list = ['cat', 'dog', 'monkey']
for x in animals_list:
    print(x)
```

Looping over tuples is the same:

```python
tup = ('dog', 'fish', 'cow', 'lion')
for animal in tup:
  print(animal) 
```

Looping over dictionary:

```python
d = {'person': 2, 'cat': 4, 'spider': 8}
for animal, legs in d.items():
    print('A {} has {} legs'.format(animal, legs))
```

```python
for key in d.keys():
  print(key)
```

```python
for val in d.values():
  print(val)
```

If you want access to the index of each element within the body of a loop, use the built-in enumerate function:

```python
animals = ['cat', 'dog', 'monkey']
for ind, animal in enumerate(animals):
    print('{}: {}'.format(ind+1 , animal))
```

To run a loop over a range of numbers, use range():

```python
for i in range(10): #range gives numbers from 0 to 9
  print(i, end=' ')
```

```python
for i in range(2,16): #range gives number from 2 to 15
  print(i, end=' ')
```

```python
for i in range(0,10,2): #range gives even numbers till 9 as step=2
  print(i, end=' ')
```

while loops can be used as:

```python
i = 10
while i > 0:
  print(i, end=' ')
  i -= 1
```

# 7 Functions

## 7.1 Normal function

Python functions are defined using the def keyword. For example:

```python
def sign(x):
    if x > 0:
        return 'positive'
    elif x < 0:
        return 'negative'
    else:
        return 'zero'

for x in [-1, 0, 1]:
    print(sign(x))
```

We will often define functions to take optional keyword arguments, like this:

```python
def hello(name, loud=False):
    if loud:
        print('HELLO, {}'.format(name.upper()))
    else:
        print('Hello, {}!'.format(name))

hello('Bob')
hello('Fred', loud=True)
```

## 7.2 Lambda function

Python has a utility called lambda function, which is a small anonymous function. It can take any number of arguments, but can only have one expression. Widely used for small purposes.

```python
x = lambda a : a + 10   #lambda arguments : expression
print(x(5))
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2)) 
```

```python
numList = [i for i in range(0,10)]
print(f"List of numbers from 1 to 10:\n{numList}")

charNum = lambda i: ord('a') + i

charDict = {chr(charNum(i)): charNum(i) for i in numList}
print(f"Dictionary of letters and numbers:\n{charDict}")
```

## 8 Classes Object Oriented Design

```python
class Rectangle():
    
    _area = -1

    def __init__(self, width, height):
        self._width = width
        self._height = height
    
    def getWidth(self):
        return self._width
    
    def getHeight(self):
        return self._height
    
    def setWidth(self, width):
        self._width = width
    
    def setHeight(self, height):
        self._height = height
        
    def getArea(self):
        return self._width * self._height

box = Rectangle(5, 10)

print("Area of the Rectangle = ", box.getArea())
box.setHeight(2)
box.setWidth(3)
print(f"New dimensions of the Rectangle: {box.getWidth()} X {box.getHeight()}")
box.getArea()
```
